```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df=pd.read_csv('../filelog.txt',sep=';') #.iloc[-100:]
df.dropna(inplace=True)
df.loc[:,'START']=pd.to_datetime(df.START)
kk=df.INTERVAL.apply(lambda x: datetime.datetime.strptime(x,"%H:%M:%S.%f"))
df.loc[:,'INTERVAL']=kk.apply(lambda x: datetime.timedelta(hours=x.hour, minutes=x.minute, seconds=x.second))
df.set_index(df.START,inplace=True)
# ...and use datetime's hour, min and sec properties to build a timedelta
#delta = timedelta(hours=t.hour, minutes=t.minute, seconds=t.second)
```


col='RA'
q_low = df[col].quantile(0.1)
q_hi  = df[col].quantile(0.9)
df = df[(df[col] < q_hi) & (df[col] > q_low)]

```python
df['deltaRA']=(df.RA.shift(periods=-1)-df.RA)*3600
df['deltaDEC']=(df.DEC.shift(periods=-1)-df.DEC)*3600
df['deltaT']=((df.START.shift(periods=-1)-df.START).dt.total_seconds())
df['speedRA']=df['deltaRA']/df['deltaT']
df['speedDEC']=df['deltaDEC']/df['deltaT']
```
